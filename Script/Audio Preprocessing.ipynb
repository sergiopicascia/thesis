##Â Extract 30s Excerpts

```python
from pydub import AudioSegment
import os
from pychorus import find_and_output_chorus
from tqdm import tqdm
import opensmile
import audiofile
import pandas as pd
from pathlib import Path
from pyAudioAnalysis import MidTermFeatures
from pyAudioAnalysis import audioBasicIO
import numpy as np
import math
```

```python
path = '/Volumes/Sergio-SSD/AudioData/'
folders = [path+str(idx) for idx in list(range(1, 2001))]
```

```python
start = 30 * 1000
end = 60 * 1000

for folder in folders:
    for file in os.listdir(folder):
        if (file.endswith('.mp3')) & (not file.startswith('.')):
            song = AudioSegment.from_file(folder+'/'+file, format='mp3')
            excerpt = song[start:end]
            excerpt.export(folder+'/'+'30s-'+file, format='mp3')
```

## Chorus Detection

```python
# Extract 10s excerpts identifying the most common pattern in song (chorus, eventually)
for folder in tqdm(folders):
    for file in os.listdir(folder):
        if (file.endswith('.mp3')) & (not file.startswith('.')) & (not file.startswith('30s')) & (not file.startswith('Chor')):
            find_and_output_chorus(folder+'/'+file, (folder+'/Chor-'+file).replace('.mp3', '.wav'), 10)
```

```python
from spleeter.separator import Separator

separator = Separator('spleeter:2stems')

for folder in folders:
    for file in os.listdir(folder):
        if (file.endswith('.mp3')) & (file.startswith('30s-')):
            separator.separate_to_file(str(folder + '/' + file), folder, codec='mp3')
        elif (file.endswith('.wav')) & (file.startswith('Chor-')):
            separator.separate_to_file(str(folder + '/' + file), folder, codec='mp3')
```

## Vocals Feature Extraction

```python
# Initialize Smile
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)

# Function for retrieving features from vocals files
def vocals_features(filepath):
    signal, sr = audiofile.read(filepath)
    features = smile.process_signal(signal, sr)
    
    features.reset_index(drop=True, inplace=True)
    features['id'] = os.path.basename(Path(filepath).parent.parent)
    features.set_index('id', inplace=True)
    
    return features
```

```python
vocals_full = []
vocals_30s = []
vocals_10s = []
```

```python
for folder in tqdm(folders):
    sub_folders = [sub_folder for sub_folder in os.listdir(folder) if os.path.isdir(folder+'/'+sub_folder)]
    
    if len(sub_folders) == 3:
        for sub_folder in sub_folders:
            file = folder+'/'+sub_folder+'/'+'vocals.mp3'
            values = vocals_features(file)

            if sub_folder.startswith('30s'):
                vocals_30s.append(values)
            elif sub_folder.startswith('Chor'):
                vocals_10s.append(values)
            else:
                vocals_full.append(values)
    else:
        for sub_folder in sub_folders:
            file = folder+'/'+sub_folder+'/'+'vocals.mp3'
            values = vocals_features(file)

            if sub_folder.startswith('30s'):
                vocals_30s.append(values)
                vocals_10s.append(values)                
            else:
                vocals_full.append(values)
```

```python
vocals_full_df = pd.concat(vocals_full)
vocals_30s_df = pd.concat(vocals_30s)
vocals_10s_df = pd.concat(vocals_10s)
```

```python
vocals_full_df.to_csv('vocals_full.csv')
vocals_30s_df.to_csv('vocals_30s.csv')
vocals_10s_df.to_csv('vocals_10s.csv')
```

## Music Feature Extraction

```python
def music_features(filepath):
    sr, signal = audioBasicIO.read_audio_file(filepath)
    signal = audioBasicIO.stereo_to_mono(signal)
    duration = math.ceil(len(signal)/sr)
    
    ltf, _, ltfn = MidTermFeatures.mid_feature_extraction(signal, sr, duration*sr, duration*sr, 0.050*sr, 0.050*sr)
    
    values = np.append(ltf, os.path.basename(Path(filepath).parent.parent))
    names = np.append(ltfn, 'id')
    
    features = pd.DataFrame(np.reshape(values, (-1, 1)).T, columns=names).set_index('id', drop=True)

    return features
```

```python
music_full = []
music_30s = []
music_10s = []
```

```python
for folder in tqdm(folders):
    sub_folders = [sub_folder for sub_folder in os.listdir(folder) if os.path.isdir(folder+'/'+sub_folder)]
    
    if len(sub_folders) == 3:
        for sub_folder in sub_folders:
            file = folder+'/'+sub_folder+'/'+'accompaniment.mp3'
            values = music_features(file)

            if sub_folder.startswith('30s'):
                music_30s.append(values)
            elif sub_folder.startswith('Chor'):
                music_10s.append(values)
            else:
                music_full.append(values)
    else:
        for sub_folder in sub_folders:
            file = folder+'/'+sub_folder+'/'+'accompaniment.mp3'
            values = music_features(file)

            if sub_folder.startswith('30s'):
                music_30s.append(values)
                music_10s.append(values)                
            else:
                music_full.append(values)
```

```python
music_full_df = pd.concat(music_full)
music_30s_df = pd.concat(music_30s)
music_10s_df = pd.concat(music_10s)
```

```python
music_full_df.to_csv('music_full.csv')
music_30s_df.to_csv('music_30s.csv')
music_10s_df.to_csv('music_10s.csv')
```
